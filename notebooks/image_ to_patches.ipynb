```python
import torch
```

```python
num_patches = 4
```

```python
a = torch.randn(3, 8, 8)
a.shape
```

```python
unfolded_h = a.unfold(dimension=1, size=num_patches, step=num_patches)
unfolded_h.shape
```

```python
unfolded_w = unfolded_h.unfold(dimension=2, size=num_patches, step=num_patches)
unfolded_w.shape
```

```python
patches = unfolded_w.reshape(
    num_patches, -1
)  # [# of patches, # of channels * patch_height * patch_width] => as each patch size is patch_height * patch_width
patches.shape
```
